### FEED FORWARD NEURAL NETWORK

```python

from graphviz import Digraph

def visualize_model():
    # Create a new Digraph with horizontal layout
    dot = Digraph(comment='Feedforward Neural Network', format='png')
    dot.attr(dpi='300', rankdir='TB')  # 'LR' for left-to-right layout

    # Define colors for the nodes
    node_colors = {
        'Input': 'lightblue',
        'H1': 'lightgreen',
        'H2': 'lightyellow',
        'H3': 'lightcoral',
        'H4': 'lightpink',
        'H5': 'lightgray',
        'Output': 'lightgoldenrod'
    }

    # Add nodes with colors and sizes
    node_sizes = {
        'Input': 10000,
        'H1': 2000,
        'H2': 500,
        'H3': 200,
        'H4': 50,
        'H5': 10,
        'Output': 3
    }
    
    for layer, color in node_colors.items():
        size = node_sizes[layer]
        dot.node(layer, label=f'{layer} Layer\n({size} units)', style='filled', fillcolor=color)

    # Add edges to represent the connections with colors
    dot.edge('Input', 'H1', label='Linear + ReLU', color='black')
    dot.edge('H1', 'H2', label='Linear + ReLU', color='black')
    dot.edge('H2', 'H3', label='Linear + ReLU', color='black')
    dot.edge('H3', 'H4', label='Linear + ReLU', color='black')
    dot.edge('H4', 'H5', label='Linear + ReLU', color='black')
    dot.edge('H5', 'Output', label='Linear', color='black')

    # Render the graph to a file
    dot.render('feedforward_nn', format='png', cleanup=True)

# Run the visualization function
visualize_model()
```

### CNN

#### AlexNET format

```python
from graphviz import Digraph

def visualize_alexnet_format():
    # Create a new Digraph with horizontal layout
    dot = Digraph(comment='SmallCNN in AlexNet Format', format='png')
    dot.attr(dpi='300', rankdir='LR')  # 'LR' for left-to-right layout

    # Define colors for the nodes
    node_colors = {
        'Input': 'lightblue',
        'Conv1': 'lightgreen',
        'BN1': 'lightyellow',
        'Pool1': 'lightcoral',
        'Conv2': 'lightgreen',
        'BN2': 'lightyellow',
        'Pool2': 'lightcoral',
        'Conv3': 'lightgreen',
        'BN3': 'lightyellow',
        'Pool3': 'lightcoral',
        'Flatten': 'lightpink',
        'FC1': 'lightgray',
        'Dropout': 'lightgoldenrod',
        'FC2': 'lightcyan',
        'Output': 'lightgoldenrod'
    }

    # Add nodes with colors and sizes, mimicking the AlexNet structure
    node_labels = {
        'Input': 'Input Layer\n(1x50x50)',
        'Conv1': 'Conv Layer 1\n(32x50x50)\n3x3 Kernel',
        'BN1': 'BatchNorm 1',
        'Pool1': 'MaxPool 1\n(32x25x25)',
        'Conv2': 'Conv Layer 2\n(64x25x25)\n3x3 Kernel',
        'BN2': 'BatchNorm 2',
        'Pool2': 'MaxPool 2\n(64x12x12)',
        'Conv3': 'Conv Layer 3\n(128x12x12)\n3x3 Kernel',
        'BN3': 'BatchNorm 3',
        'Pool3': 'MaxPool 3\n(128x6x6)',
        'Flatten': 'Flatten\n(128*6*6)',
        'FC1': 'Fully Connected 1\n(512 units)',
        'Dropout': 'Dropout\n(50%)',
        'FC2': 'Fully Connected 2\n(3 units)',
        'Output': 'Output Layer\n(3 units)'
    }

    for layer, label in node_labels.items():
        color = node_colors[layer]
        dot.node(layer, label=label, style='filled', fillcolor=color)

    # Add edges to represent the connections in AlexNet style
    dot.edge('Input', 'Conv1', color='black')
    dot.edge('Conv1', 'BN1', color='black')
    dot.edge('BN1', 'Pool1', color='black')
    dot.edge('Pool1', 'Conv2', color='black')
    dot.edge('Conv2', 'BN2', color='black')
    dot.edge('BN2', 'Pool2', color='black')
    dot.edge('Pool2', 'Conv3', color='black')
    dot.edge('Conv3', 'BN3', color='black')
    dot.edge('BN3', 'Pool3', color='black')
    dot.edge('Pool3', 'Flatten', color='black')
    dot.edge('Flatten', 'FC1', color='black')
    dot.edge('FC1', 'Dropout', color='black')
    dot.edge('Dropout', 'FC2', color='black')
    dot.edge('FC2', 'Output', color='black')

    # Render the graph to a file
    dot.render('small_cnn_alexnet_format', format='png', cleanup=True)

# Run the visualization function
visualize_alexnet_format()
```

### NEW FORMAT


```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def add_conv_layer(ax, position, width, height, text):
    rect = patches.Rectangle(position, width, height, linewidth=2, edgecolor='black', facecolor='sandybrown')
    ax.add_patch(rect)
    ax.text(position[0] + width / 2, position[1] + height / 2, text, 
            ha='center', va='center', fontsize=12, color='black')

def add_pooling_layer(ax, position, width, height, text):
    rect = patches.Rectangle(position, width, height, linewidth=2, edgecolor='black', facecolor='lightcoral')
    ax.add_patch(rect)
    ax.text(position[0] + width / 2, position[1] + height / 2, text, 
            ha='center', va='center', fontsize=12, color='black')

def add_fc_layer(ax, position, width, height, text):
    rect = patches.Rectangle(position, width, height, linewidth=2, edgecolor='black', facecolor='lightgray')
    ax.add_patch(rect)
    ax.text(position[0] + width / 2, position[1] + height / 2, text, 
            ha='center', va='center', fontsize=12, color='black')

def visualize_smallcnn_alexnet():
    fig, ax = plt.subplots(figsize=(15, 7))
    ax.set_xlim(0, 14)
    ax.set_ylim(0, 5)
    ax.axis('off')

    # Input Image
    ax.text(0.5, 4, 'Input\n(1x50x50)', ha='center', va='center', fontsize=12, color='black')

    # Convolutional and Pooling Layers
    add_conv_layer(ax, (2, 3), 1.5, 1, 'Conv1\n(32x50x50)')
    add_pooling_layer(ax, (3.5, 3), 1.5, 1, 'MaxPool\n(32x25x25)')

    add_conv_layer(ax, (5, 3), 1.5, 1, 'Conv2\n(64x25x25)')
    add_pooling_layer(ax, (6.5, 3), 1.5, 1, 'MaxPool\n(64x12x12)')

    add_conv_layer(ax, (8, 3), 1.5, 1, 'Conv3\n(128x12x12)')
    add_pooling_layer(ax, (9.5, 3), 1.5, 1, 'MaxPool\n(128x6x6)')

    # Fully Connected Layers
    add_fc_layer(ax, (11, 2.5), 1.5, 2, 'FC1\n(512 units)')
    add_fc_layer(ax, (12.5, 2.5), 1.5, 2, 'FC2\n(3 units)')

    plt.sh()

visualize_smallcnn_alexnet()
```
